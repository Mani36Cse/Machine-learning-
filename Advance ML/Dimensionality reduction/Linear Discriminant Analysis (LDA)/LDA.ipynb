```python
import pandas as pd
from LDA import lda 
ld=lda()
```

```python
data_set=pd.read_csv("Wine.csv")
data_set
```

```python
#No categorical data in dataset so no need to do get_dummies
```

```python
#input and output split
Indep_x=data_set.drop("Customer_Segment",axis=1)
Depend_y=data_set["Customer_Segment"]
```

```python
#Split the train and test data
x_train,x_test,y_train,y_test=ld.split_scaler(Indep_x,Depend_y)
```

```python
#Give input and output to LDA
#number of components (<= number of classes -1)
x_train,x_test,explained_variance=ld.LDA_input(x_train,x_test,y_train,2)
```

```python
explained_variance
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1, len(explained_variance)+1), 
         explained_variance, marker='o')
plt.title('Explained Variance by Each Component')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.grid(True)
plt.show()
```

```python
#After the PCA we get the liear pattern data, PCA will extract the linear data, using that data we can perfor the Supervise --- Classification
acclog=[]
accsvml=[]
accknn=[]
accdes=[]
accrf=[]
```

```python
classifier,Accuracy,report,x_test,y_test,cm=ld.Logistic(x_train,y_train,x_test,y_test)
acclog.append(Accuracy)

classifier,Accuracy,report,x_test,y_test,cm=ld.svm_linear(x_train,y_train,x_test,y_test)  
accsvml.append(Accuracy)

classifier,Accuracy,report,x_test,y_test,cm=ld.knn(x_train,y_train,x_test,y_test)  
accknn.append(Accuracy)

classifier,Accuracy,report,x_test,y_test,cm=ld.Decision(x_train,y_train,x_test,y_test)  
accdes.append(Accuracy)
    
classifier,Accuracy,report,x_test,y_test,cm=ld.random(x_train,y_train,x_test,y_test)  
accrf.append(Accuracy)
```

```python
result=ld.LDA_Classification(acclog,accsvml,accknn,accdes,accrf)
```
