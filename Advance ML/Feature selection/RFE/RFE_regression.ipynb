```python
import pandas as pf
from RFE_regression import RFE_regression
rfe_re=RFE_regression()
```

```python
data_set=pf.read_csv("prep.csv")
data_set
```

```python
data_set.isna().sum()
#No null values found
```

```python
#data preprocessing
dataset=pf.get_dummies(data_set,drop_first=True)
dataset
```

```python
#input and output split
inde_x=dataset.drop("classification_yes",axis=1)
depe_y=dataset["classification_yes"]
```

```python
#Give input and output to RFE best algorithm, 5 no of cloumns accept 
value_x,column=rfe_re.rfe_feature(inde_x,depe_y,5)
```

```python
acclin=[]
accsvml=[]
accsvmnl=[]
accdes=[]
accrf=[]
```

```python
#In Vlaue_x it store the all RFE model in list data type, so we have that list data one by one to predict the value
#Using the loop we get the model one by one, one model came and went to train and test, check in other models and give result
for i in value_x:
    x_train,x_test,y_train,y_test=rfe_re.split_scaler(i,depe_y)
    r2_lin=rfe_re.Linear(x_train,y_train,x_test,y_test)
    acclin.append(r2_lin)
    
    r2_sl=rfe_re.svm_linear(x_train,y_train,x_test,y_test)    
    accsvml.append(r2_sl)
    
    r2_NL=rfe_re.svm_NL(x_train,y_train,x_test,y_test)
    accsvmnl.append(r2_NL)
    
    r2_d=rfe_re.Decision(x_train,y_train,x_test,y_test)
    accdes.append(r2_d)
    
    r2_r=rfe_re.random(x_train,y_train,x_test,y_test)
    accrf.append(r2_r)
```

```python
result=rfe_re.rfe_regression(acclin,accsvml,accdes,accrf)
```
