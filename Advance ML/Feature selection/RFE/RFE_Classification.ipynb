```python
import pandas as pd
from RFE_classification import RFE_Classification
RC=RFE_Classification()
```

```python
data_set=pd.read_csv("prep.csv")
#check the sample data
data_set.head(10)
```

```python
#check rows and columns 
data_set.shape
```

```python
data_set.isna().sum()
#No null values found
```

```python
#data preprocessing
dataset=pd.get_dummies(data_set,drop_first=True)
dataset.head(5)
```

```python
#input and output split
inde_x=dataset.drop("classification_yes",axis=1)
depe_y=dataset["classification_yes"]
```

```python
#Give input and output to RFE best algorithm, 5 no of cloumns accept 
value_x,column=RC.rfe_features(inde_x,depe_y,5)
```

```python
print(column)
#These input columns are selected by the RFE 
```

```python
acclog=[]
accsvc=[]
accdes=[]
accrf=[]
```

```python
#In value_x have more then one RFE result in list format, so give it loop get the result one by one.
#one model came and went to train and test, check in other models and give result
for i in value_x:
    x_train, x_test, y_train, y_test=RC.split_scaler(i,depe_y)
    
    classifier,Accuracy,report,x_test,y_test,cm=RC.Logistic(x_train,y_train,x_test,y_test)
    acclog.append(Accuracy)
    
    classifier,Accuracy,report,x_test,y_test,cm=RC.svc_linear(x_train,y_train,x_test,y_test)  
    accsvc.append(Accuracy)
    
    classifier,Accuracy,report,x_test,y_test,cm=RC.Decision(x_train,y_train,x_test,y_test)  
    accdes.append(Accuracy)
    
    classifier,Accuracy,report,x_test,y_test,cm=RC.random(x_train,y_train,x_test,y_test)  
    accrf.append(Accuracy)
```

```python
#It store the 4 type of algorithms scores
accsvc
```

```python
result=RC.rfe_classification(acclog,accsvc,accdes,accrf)
```
