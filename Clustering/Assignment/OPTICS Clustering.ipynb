```python
import pandas as pd
```

### 1. Data collection

```python
Data=pd.read_csv("Mall_Customers.csv")
Data
```

### 2.Split the input data

```python
X=Data.iloc[:,3:5].values
X
```

### 3. Model creation

```python
from sklearn.cluster import OPTICS
OP=OPTICS(min_samples=12)
y_mean=OP.fit_predict(X)
y_mean
```

```python
supervised=Data
supervised["Cluster_group"]=y_mean
supervised
```

### 4. Final result

```python
import seaborn as sea
out=sea.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], fit_reg=False)
```

### 5.Evaluation Metric

```python
from sklearn.metrics import davies_bouldin_score
result=davies_bouldin_score(X, y_mean)
result
```

```python
#In this predict lable value have more -1 means it was noise or outlier.
```

```python
#Good clustering: DBI typically ranges between 0.5 to 2.0.

#Poor clustering: DBI scores above 3 or 4 often mean:
#Clusters are not well-separated.
#Clusters are not compact (spread out).
#There may be too much noise or too many outliers.
```
