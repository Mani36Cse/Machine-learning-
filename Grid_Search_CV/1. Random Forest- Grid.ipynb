```python
import pandas as pd
```

## 1.Data collection

```python
Data=pd.read_csv("50_Startups.csv")
Data
```

```python
Data=pd.get_dummies(Data, drop_first=True)
Data
```

```python
Data.columns
```

## 2.Input and Output split

```python
Independent=Data[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
Independent
```

```python
Dependent=Data[['Profit']]
Dependent
```

## 3. Split the train and test data

```python
#Here we are using Gird Search CV (cross validation) so no need to use train and test split, Gird will do that all the method.
```

## 4. Model selection

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor

Kfold=KFold(n_splits=5,shuffle=True, random_state=42)

param_grid={'n_estimators': [10,30,50,100], 'criterion': ['squared_error', 'absolute_error', 'friedman_mse', 'poisson']}

#We give refit true value so it will take the best model from the final resulr if it in false it will the final result which is the end of the result in this it will take n_estimators : 100 and criterion : poisson those value are in the end. 
grid=GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=Kfold, refit=True, verbose=3, n_jobs=-1)

grid.fit(Independent, Dependent)
```

## 5.Evalutation Metric

```python
#since we are using GridSearchCV so it will do Evalution the metric and give the best model
```

```python
re=grid.cv_results_

# print the best parametar after tuning 
print("The Score value for the best model {}".format(grid.best_params_),"Score :",grid.best_score_)
```

```python
table=pd.DataFrame.from_dict(re)
table
```
