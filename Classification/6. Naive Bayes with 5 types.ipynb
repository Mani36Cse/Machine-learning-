```python
import pandas as pd
import numpy as np
```

### 1. Data collection

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset
```

```python
data=pd.get_dummies(dataset,drop_first=True)
```

```python
data
```

```python
#no use for User ID so we drop that column in the table data
data=data.drop("User ID", axis=1)
data
```

```python
data["Purchased"].value_counts()
```

### 2.Split the input and output

```python
independent=data[["Age","EstimatedSalary","Gender_Male"]]
independent
```

```python
dependent=data[["Purchased"]]
dependent
```

### 3.Split the train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent, test_size=0.30,random_state=0)
```

```python
x_train
```

### 4. Model creation and 5.Evaluation Metric

```python
#we have 5 types in navie_bayes algorithm 
```

```python
#BernoulliNB type 
from sklearn.naive_bayes import BernoulliNB
classifierB=BernoulliNB()
classifierB.fit(x_train,y_train)

y_pred = classifierB.predict(x_test)
#Evalute the matricdatas

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
#Report 

from sklearn.metrics import classification_report
clf=classification_report(y_test,y_pred)

print(cm)
print(clf)
```

```python
#CategoricalNB
from sklearn.naive_bayes import CategoricalNB
classifierC=CategoricalNB()
classifierC.fit(x_train,y_train)

y_pred = classifierC.predict(x_test)
#Evalute the matrics

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
#Report 

from sklearn.metrics import classification_report
clf=classification_report(y_test,y_pred)

print(cm)
print(clf)
```

```python
#ComplementNB
from sklearn.naive_bayes import ComplementNB
classifierCo=ComplementNB()
classifierCo.fit(x_train,y_train)

y_pred = classifierCo.predict(x_test)
#Evalute the matrics

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
#Report 

from sklearn.metrics import classification_report
clf=classification_report(y_test,y_pred)

print(cm)
print(clf)
```

```python
#GaussianNB
from sklearn.naive_bayes import GaussianNB
classifierG=GaussianNB()
classifierG.fit(x_train,y_train)

y_pred = classifierG.predict(x_test)
#Evalute the matrics

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
#Report 

from sklearn.metrics import classification_report
clf=classification_report(y_test,y_pred)

print(cm)
print(clf)
```

```python
#MultinomialNB
from sklearn.naive_bayes import MultinomialNB
classifierM=MultinomialNB()
classifierM.fit(x_train,y_train)

y_pred = classifierM.predict(x_test)
#Evalute the matrics

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
#Report 

from sklearn.metrics import classification_report
clf=classification_report(y_test,y_pred)

print(cm)
print(clf)
```
