```python
import pandas as pd
```

```python
Data=pd.read_csv("Social_Network_Ads.csv")
Data
```

```python
data=pd.get_dummies(Data,drop_first=True)
data
```

```python
data=data.drop("User ID",axis=1)
data
```

```python
data["Purchased"].value_counts()
#to check the give data balanced or imbalanced 
```

```python
Independent=data[["Age","EstimatedSalary","Gender_Male"]]
Independent
```

```python
Dependent=data[["Purchased"]]
Dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(Independent, Dependent, test_size=0.30, random_state=0)
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x_train=st.fit_transform(x_train)
x_test=st.transform(x_test)
```

```python
x_test
```

```python
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import GridSearchCV

param_grid ={'criterion':['gini','entropy'],
              'max_features': ['auto','sqrt','log2'],
              'splitter':['best','random']}

grid=GridSearchCV(DecisionTreeClassifier(), param_grid, refit= True, verbose=3, n_jobs=-1, scoring="f1_weighted")

grid.fit(x_train,y_train)
```

```python
print(grid.best_params_)
#Using this we can see the best model atfer tuning
re=grid.cv_results_
#save the result and we can see this table method using pandas
#print(re)
```

```python
y_pred=grid.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
#74 correctly predict as not purchased and 5 wrongly predict as not purchased
#7 correctly predict as purchased and 35 wrongly predict as purchased 

#5 is the error 1 and 7 is error 2, error-1 is less then error-2 so result will be good
```

```python
from sklearn.metrics import classification_report
clr=classification_report(y_test,y_pred)
print(clr)
```

```python
#to see the specific vales in the confusion matrics, in average what ever we pass the metrics it will that value
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,y_pred,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

⚖️ When to use what?
Scenario	Recommended average
Binary classification	'binary' (default if 2 classes)
Multiclass, balanced classes	'macro'
Multiclass, imbalanced classes	'weighted'
You want total accuracy-like score	'micro'
Multilabel classification	'samples' or 'micro'
