```python
import pandas as pd
import numpy as np
```

### 1. Data collection

```python
Data_set=pd.read_csv("Social_Network_Ads.csv")
Data_set
```

```python
Data=pd.get_dummies(Data_set,drop_first=True)
Data
```

```python
Data=Data.drop("User ID",axis=1)
Data
```

```python
Data["Purchased"].value_counts()
```

### 2.Split the input and output

```python
independent=Data[["Age","EstimatedSalary","Gender_Male"]]
independent
```

```python
dependent=Data[["Purchased"]]
dependent
```

### 3.Split the train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
x_test
```

### 4.Model crearion

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
param_grid = {'solver':['newton-cg', 'lbfgs', 'liblinear', 'saga'],'penalty':['l1','l2']}

grid=GridSearchCV(LogisticRegression(), param_grid, refit=True, verbose=3, n_jobs=-1, scoring='f1_weighted')
#scoring w mention which metrics we need to get default value of scoring is For classifiers: usually accuracy and regressors: usually R² score
#we select the f1_weighted so it will give the high score value on this f1_weighted. we can select other metrics also

grid.fit(x_train,y_train)
```

### 5.Evaluation Metric

```python
print(grid.best_params_)
#Using this we can see the best model atfer tuning
re=grid.cv_results_
#save the result and we can see this table method using pandas
#print(re)
```

```python
y_pred=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
#0 correctly predict as not purchased and 79 wrongly predict as not purchased
#41 correctly predict as purchased and 0 wrongly predict as purchased 
```

```python
#print the classifier report
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
print(cr)
```

```python
#to see the specific vales in the confusion matrics, in average what ever we pass the metrics it will that value
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,y_pred,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

⚖️ When to use what?
Scenario	Recommended average
Binary classification	'binary' (default if 2 classes)
Multiclass, balanced classes	'macro'
Multiclass, imbalanced classes	'weighted'
You want total accuracy-like score	'micro'
Multilabel classification	'samples' or 'micro'
