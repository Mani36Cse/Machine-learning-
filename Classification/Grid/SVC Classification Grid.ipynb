```python
import pandas as ps
import numpy as np
```

### 1. Data collection

```python
data_set=ps.read_csv("Social_Network_Ads.csv")
```

```python
data_set
```

```python
Data=ps.get_dummies(data_set,drop_first=True)
```

```python
Data
```

```python
Data=Data.drop("User ID", axis=1)
Data
```

```python
Data["Purchased"].value_counts()
```

### 2.Split the input and output

```python
independent=Data[["Age","EstimatedSalary","Gender_Male"]]
independent
```

```python
dependent=Data[["Purchased"]]
dependent
```

### 3.Split the train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x_train=st.fit_transform(x_train)
x_test=st.transform(x_test)
```

```python
x_train
```

### 4.Model crearion

```python
from sklearn.svm import SVC

from sklearn.model_selection import GridSearchCV

param_grid ={'kernel':['linear','rbf','poly','sigmoid'],
             'gamma':['auto','scale'],
             'C':[10,100,1000,2000,3000]}

grid=GridSearchCV(SVC(probability=True), param_grid, refit= True, verbose=3, n_jobs=-1, scoring="f1_weighted")

grid.fit(x_train,y_train)
```

```python
y_pred=grid.predict(x_test)
```

```python
y_pred
```

```python
print(grid.best_params_)
#Using this we can see the best model atfer tuning
re=grid.cv_results_
#save the result and we can see this table method using pandas
#print(re)
```

### 5.Evaluation Metric

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,y_pred)
print(clf_report)
```

```python
#to see the specific vales in the confusion matrics, in average what ever we pass the metrics it will that value
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,y_pred,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

⚖️ When to use what?
Scenario	Recommended average
Binary classification	'binary' (default if 2 classes)
Multiclass, balanced classes	'macro'
Multiclass, imbalanced classes	'weighted'
You want total accuracy-like score	'micro'
Multilabel classification	'samples' or 'micro'
