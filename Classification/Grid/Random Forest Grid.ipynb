```python
import pandas as ps
```

### 1.Data collection

```python
data_set=ps.read_csv("Social_Network_Ads.csv")
```

```python
data_set
```

```python
data=ps.get_dummies(data_set,drop_first=True)
data
```

```python
data=data.drop("User ID", axis=1)
data
```

```python
data["Purchased"].value_counts()
```

### 2.Split the input and output

```python
independent=data[["Age","EstimatedSalary","Gender_Male"]]
independent
```

```python
dependent=data[["Purchased"]]
dependent
```

### 3.Split the train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
sr=StandardScaler()
x_train=sr.fit_transform(x_train)
x_test=sr.transform(x_test)
```

```python
x_train
```

### 4.Model crearion

```python
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import GridSearchCV

par_grid={'criterion':['gini','entropy'],
              'max_features': ['auto','sqrt','log2'],
              'n_estimators':[10,100]} 

grid=GridSearchCV(RandomForestClassifier(), par_grid, refit=True, verbose=3, n_jobs=-1, scoring="f1_weighted")

grid.fit(x_train,y_train)
```

```python
y_pred=grid.predict(x_test)
```

```python
y_pred
```

```python
print(grid.best_params_)
#Using this we can see the best model atfer tuning
re=grid.cv_results_
#save the result and we can see this table method using pandas
#print(re)
```

### 5.Evaluation Metric

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,y_pred)
```

```python
print(clf_report)
```

```python
#to see the specific vales in the confusion matrics, in average what ever we pass the metrics it will that value
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,y_pred,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```
