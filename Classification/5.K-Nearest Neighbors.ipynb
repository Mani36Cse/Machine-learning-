```python
import pandas as pd
import numpy as  np
```

### 1. Data collection

```python
Data_set=pd.read_csv("Social_Network_Ads.csv")
Data_set
```

```python
Data=pd.get_dummies(Data_set,drop_first=True)
Data
```

```python
Data=Data.drop("User ID",axis=1)
Data
```

```python
Data["Purchased"].value_counts()
```

### 2.Split the input and output

```python
independent=Data[["Age","EstimatedSalary","Gender_Male"]]
independent
```

```python
dependent=Data[["Purchased"]]
dependent
```

### 3.Split the train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
x_train
```

### 4. Model creation

```python
from sklearn.neighbors import KNeighborsClassifier
#here n_neighbors it a k value which is represent the how many neighbors we have to take in odd number wise we can increase the number if we not getting correct evalution value
Classifier=KNeighborsClassifier(n_neighbors=7, metric = 'minkowski', p = 2)
Classifier.fit(x_train,y_train)
```

```python
y_pred=Classifier.predict(x_test)
y_pred
```

### 5.Evaluation Metric

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
clf=classification_report(y_test,y_pred)
```

```python
print(clf)
```
