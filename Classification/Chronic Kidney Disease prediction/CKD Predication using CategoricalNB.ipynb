### 1. Data collection

```python
import pandas as pd
```

```python
Data_set=pd.read_csv("CKD.csv")
Data_set
```

### 2.Split the input and output

#Instead of get_dummies we are using this Skleran one hot encoding method in get dummies it will possiable to loss some categorical data to prevent this we are using OrdinalEncoder
from sklearn.preprocessing import OrdinalEncoder 

encoder=OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=0)

#this method is advance method for split the input and output x code will remove the output column and store the remaining data column, we can select the cloumn based on the requirement
#y it store the output cloumn data
x=Data_set.drop("classification", axis=1)

y

```python
#other method in one hot encoding 
from sklearn.preprocessing import OneHotEncoder

encode=OneHotEncoder(handle_unknown='ignore',sparse=False)

x=Data_set.drop("classification", axis=1)

y=Data_set["classification"]
```

```python
Data_set["classification"].value_counts()
#Data set was imbalance data
```

### 3.Split the train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.20, random_state=42, stratify=y)
```

```python
x_train
```

```python
print(x_train.head())
```

```python
x_train_en=encode.fit_transform(x_train)
x_test_en=encode.transform(x_test)
```

```python
print(x_test_en)
```

### 4.Model crearion

```python
#CategoricalNB

from sklearn.naive_bayes import CategoricalNB

from sklearn.model_selection import GridSearchCV

param_grid = {'alpha': [0.1, 1.0]}

grid=GridSearchCV(CategoricalNB(), param_grid, refit=True, verbose=3, n_jobs=-1, scoring='f1_weighted')

grid.fit(x_train_en,y_train)


y_pred_1=grid.predict(x_test_en)

from sklearn.metrics import confusion_matrix
cm_1=confusion_matrix(y_test,y_pred_1)

from sklearn.metrics import classification_report
cl=classification_report(y_test,y_pred_1)

from sklearn.metrics import roc_auc_score
roc=roc_auc_score(y_test,grid.predict_proba(x_test_en)[:,-1])

print("---> Naive Bayes - CategoricalNB  using Grid <---")
print()
print("Best value in Grid search :",grid.best_params_)
print()
print("Validaction values")
print(cm_1)
print()
print("Classifiaction reports")
print(cl)
print("roc and auc value :",roc)
```

```python
from sklearn.naive_bayes import CategoricalNB
classifierC=CategoricalNB()
classifierC.fit(x_train_en,y_train)

y_pred = classifierC.predict(x_test_en)
#Evalute the matrics

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
#Report 

from sklearn.metrics import classification_report
clf=classification_report(y_test,y_pred)

print(cm)
print(clf)
```
