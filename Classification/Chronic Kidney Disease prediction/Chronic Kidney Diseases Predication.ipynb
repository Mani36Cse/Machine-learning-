### 1. Data collection

```python
import pandas as pd
```

```python
Data_set=pd.read_csv("CKD.csv")
Data_set
```

```python
Data=pd.get_dummies(Data_set,drop_first=True)
Data
```

```python
#above data we can't able to find all columns so we using this code to view all columns
sample=pd.set_option('display.max_columns',None)
print(Data.head())
```

### 2.Split the input and output

```python
Independent=Data[["age","bp","al","su","bgr","bu","sc","sod","pot","hrmo","pcv","wc","rc","sg_b","sg_c","sg_d","sg_e","rbc_normal","pc_normal","pcc_present","ba_present","htn_yes","dm_yes","cad_yes","appet_yes","pe_yes","ane_yes"]]
Independent
```

```python
Dependent=Data[["classification_yes"]]
Dependent
```

```python
Data["classification_yes"].value_counts()
#Data set was imbalance data
```

### 3.Split the train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(Independent,Dependent, test_size=0.30, random_state=0)
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
x_test
```

### 4.Model crearion

```python
#LogisticRegression
from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import GridSearchCV

param_grid = {'solver':['newton-cg', 'lbfgs', 'liblinear', 'saga'],'penalty':['l1','l2'],
                "multi_class":['auto', 'ovr', 'multinomial']}

grid=GridSearchCV(LogisticRegression(), param_grid, refit=True, verbose=3, n_jobs=-1, scoring='f1_weighted')

grid.fit(x_train,y_train)


y_pred=grid.predict(x_test)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)

from sklearn.metrics import classification_report
clf=classification_report(y_test,y_pred)

from sklearn.metrics import roc_auc_score
roc=roc_auc_score(y_test,grid.predict_proba(x_test)[:,-1])

print("---> LogisticRegression using Grid <---")
print()
print("Best value in Grid search :",grid.best_params_)
print()
print("Validaction values")
print(cm)
print()
print("Classifiaction reports")
print(clf)
print("roc and auc value :",roc)
```
