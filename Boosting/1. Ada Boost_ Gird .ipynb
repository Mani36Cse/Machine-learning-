```python
import pandas as pd
```

## 1. Data collection

```python
data=pd.read_csv("insurance_pre.csv")
data
```

```python
data_set=pd.get_dummies(data, drop_first=True)
data_set
```

```python
data_set.columns
```

## 2. Input and Output split

```python
Independent=data_set[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
Independent
```

```python
Dependent=data_set[['charges']]
Dependent
```

## 3.Split train and test data

```python
#Here we are using Gird Search CV (cross validation) so no need to use train and test split, Gird will do that all the method.
```

## 4. Model creatio

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostRegressor

p_grid={'n_estimators':[10,50,100],'loss':['linear', 'square', 'exponential']}

grid=GridSearchCV(AdaBoostRegressor(),p_grid, refit=True, verbose=3, n_jobs=-1)

#fiting the model in grid
grid.fit(Independent, Dependent)
#The data is split into 5 parts (folds), and the model is trained 5 times, each time using 4 folds for training and 1 fold for validation.
#9 models with different settings that are being tested. we have 2 parmeter each have 3 so 3x3=9 
#5 rows and 9 columns = 45 fits 
```

## 5. Evalution metric

```python
#since we are using GridSearchCV so it will do Evalution the metric and give the best model
```

```python
re=grid.cv_results_
print("The Score value for the best model {}".format(grid.best_params_),"Score :",grid.best_score_)
```
