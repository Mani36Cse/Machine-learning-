```python
import pandas as ps
```

## 1.Data collection

```python
Data_set=ps.read_csv("50_Startups.csv")
```

```python
Data_set
```

```python
Data_set=ps.get_dummies(Data_set,drop_first=True)
Data_set
```

```python
Data_set.columns
```

## 2. Input & Output Split

```python
Independent=Data_set[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
Independent
```

```python
Dependent=Data_set[['Profit']]
Dependent
```

## 3. Split the train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(Independent, Dependent, test_size=0.30, random_state=0)
```

## 4. Model creation

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion="absolute_error",splitter='random')
regressor=regressor.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(x_test)
```

```python
y_pred
```

## 5. Evalutation Metric 

```python
from sklearn.metrics import r2_score
r_value=r2_score(y_test,y_pred)
```

```python
r_value
```

## 6. Save the best model

```python
import pickle 
file="Final_model_DT.sav"
```

```python
pickle.dump(regressor,open(file,'wb'))
```
